## Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def contar_letra(cadena): #Diccionario vacío para guardar cada letra y su número de repeticiones
    frecuencias = {}
    for letra in cadena.lower():# .lower() convierte todo a minúsculas
        if letra != " ": # No contamos los espacios
            if letra in frecuencias:
                frecuencias[letra] += 1 #Si la letra ya está +1
            else:
                frecuencias[letra] = 1  # Si la letra aparece +1

    for letra, cantidad in frecuencias.items(): # cantidad=1=vez cantidad=2=veces
        if cantidad == 1:
            print(f"La letra || {letra} || se repite {cantidad} vez.")
        else:
            print(f"La letra || {letra} || se repite {cantidad} veces.")

frase = "Primer ejercicio de la katas"
print(contar_letra(frase))


```

## Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
def buscar_palabras_con_objetivo(lista_palabras, palabra_objetivo):
    palabras_encontradas = []#Lista vacía guardaremos las coincidencias
    
    for palabra in lista_palabras:# Recorremos cada palabra de la lista original
        if palabra_objetivo.lower() in palabra.lower():  # usamos lower() para ignorar mayúsculas/minúsculas
            palabras_encontradas.append(palabra) # Añadimos palabra a la lista de resultados
    
    return palabras_encontradas# Devolvemos la lista final 


lista = ["manzana", "banana", "naranja", "sandía", "mandarina"]
objetivo = "man"

resultado = buscar_palabras_con_objetivo(lista, objetivo)
#.join pasa lista a cadena
print(f"Palabras que contienen || {objetivo} ||: {', '.join(resultado)}")
```

## Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado (por defecto 5). La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota_aprobado. Si es así, el estado será "aprobado"; de lo contrario, "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
def calcular_media(nota, nota_aprobado=5): 
    sum_total = sum(notas) #calculamos la suma total de las notas sum()
    cantidad = len(notas) #len() saca la cantidad de las notas
    media = sum_total / cantidad # suma entre cantidad de notas nos da la media

    if media >= nota_aprobado:
        estado = "Enhorabuena has aprobado."
    else:
        estado = "Lo sentimos has suspendido."

    return (media, estado) #tupla media y nota

notas = [4,6,10,10,6,1]

resultado = calcular_media(notas)

print(f"Notas del alumno: {notas}")

print(f"Media: {resultado[0]:.2f} = Estado: {resultado[1]}")
```

Kata 5
## Escribe una función que calcule el factorial de un número de manera recursiva.

```python
def factorial(n):
    if n == 0 or n == 1: #si no es cero o uno es siempre uno
        return 1
    else:
        return n * factorial(n - 1) #recuesividad multiplicamos n por el factorial de (n -1)

num = 5

resultado = factorial(num)

print(f"El factorial del {num} es {resultado}")
```

## Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una excepción personalizada y maneja el error adecuadamente.

```python
class ListaVaciaError(Exception): #La clase coje la excepción si la lista está vacía
    pass
def calcular_promedio(lista_numeros):
    if len(lista_numeros) == 0:
        raise ListaVaciaError("La lista está vacía. No se puede calcular el promedio.")
    
    promedio = sum(lista_numeros) / len(lista_numeros)
    return promedio


# excepciones
try:
    numeros = [4, 7, 9, 10, 6]   
    resultado = calcular_promedio(numeros)
    print(f"El promedio es: {resultado:.2f}")


except ListaVaciaError as e:
    print("Error:", e)


except Exception as e:
    print("Ha ocurrido un error inesperado:", e)

finally:
    print("Fin del programa.")
```

Kata 27
## Crea una función que calcule el promedio de una lista de números.

```python
def promedio(lista_numeros):
    if len(lista_numeros) == 0:#lista vacia
        return "Error: la lista está vacía."
    
    suma = sum(lista_numeros)#sumamos los números de la lista
    
    media = suma / len(lista_numeros)#sacamos la media

    return media

numeros = [15, 70, 19, 93, 698]
resultado = promedio(numeros)

print("Listado de números:", numeros)
print(f"El promedio es: {resultado:.2f}")
```

Kata 29
## Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el carácter '#' excepto los últimos cuatro.

```python
def enmascarar_variable(valor):
    texto = str(valor)#cadena a texto

    if len(texto) <= 4:#menos de cuatro caracteres sin #
        return texto

    texto_enmascarado = '#' * (len(texto) - 4) + texto[-4:]#más de cuatro sustituimos por #

    return texto_enmascarado


datos = ["123456789", "abcd1234", "9876", "abc"]
for d in datos:
    print(f"Original: {d}  -->  Enmascarado: {enmascarar_variable(d)}")
```

Kata 30
## Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

```python
def anagramas(palabra1, palabra2):

    palabra1 = palabra1.lower()#convertimos a minusculas
    palabra2 = palabra2.lower()

    palabra1 = palabra1.replace(" ", "")#quitamos espacios y huecos
    palabra2 = palabra2.replace(" ", "")

    return sorted(palabra1) == sorted(palabra2) # Ordenamos y comparamos a ver si son anagramas

print(anagramas("Roma", "Amor"))       # True
print(anagramas("perro", "gato"))      # False
print(anagramas("Listen", "Silent"))   # True
print(anagramas("Lápiz", "ziláp"))     # True (si quitamos tildes)
```

```python
class Arbol:
    def __init__(self):
        self.tronco = 1          # Inicio del tronco
        self.ramas = []          

    def crecer_tronco(self):
        self.tronco += 1 # sumamos uno pra aumnentar la longitud

    def nueva_rama(self):
        self.ramas.append(1) #rama nueva con 1 de ongitud inicial 

    def crecer_ramas(self):
        # Usamos un bucle para recorrer cada rama y sumarle 1
        self.ramas = [rama + 1 for rama in self.ramas]

    def quitar_rama(self, posicion): #para quitar ramas desde la posiión que damos
        if 1 <= posicion <= len(self.ramas):
            del self.ramas[posicion - 1]
        else:
            print("La posición y o la rama no son válidas")

    def info_arbol(self):
        print("=== INFORMACIÓN DEL ÁRBOL ===")
        print(f"Longitud del tronco: {self.tronco}")
        print(f"Número de ramas: {len(self.ramas)}")
        if self.ramas:
            print(f"Longitudes de las ramas: {self.ramas}")
        else:
            print("Este árbol todavía no tiene ramas.")



mi_arbol = Arbol()

mi_arbol.crecer_tronco()

mi_arbol.nueva_rama()

mi_arbol.crecer_ramas()

mi_arbol.nueva_rama()
mi_arbol.nueva_rama()

mi_arbol.quitar_rama(2)

mi_arbol.info_arbol()
```

Crea la clase UsuarioBanco
Representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta corriente.
Métodos: retirar_dinero, transferir_dinero, agregar_dinero.
Código a seguir:
Inicializar un usuario con nombre, saldo y un indicador (True o False) de cuenta corriente.
Implementar retirar_dinero para sustraer dinero del saldo, lanzando un error si no es posible.
Implementar transferir_dinero para transferir dinero desde otro usuario, lanzando un error en caso de fallo.
Implementar agregar_dinero para aumentar el saldo del usuario.
Caso de uso:
        a. Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
        b. Agregar 20 unidades al saldo de Bob.
        c. Transferir 80 unidades de Bob a Alicia.
        d. Retirar 50 unidades del saldo de Alicia.

```python
class UsuarioBanco:
    def __init__(self, nombre, saldo, tiene_cuenta_corriente): #iniciamos datos
        self.nombre = nombre
        self.saldo = saldo
        self.tiene_cuenta_corriente = tiene_cuenta_corriente

    def retirar_dinero(self, cantidad):#sacar dinero y mensajes segun datos
        if cantidad <= 0:
            raise ValueError(" Cantidad no válida, tienes que retirar una cantidad positiva.")
        if cantidad > self.saldo:
            raise ValueError(f"{self.nombre} no tienes saldo suficiente.")
        self.saldo -= cantidad
        print(f"{self.nombre} ha retirado {cantidad}€. Te queda actualmente: {self.saldo}€ de saldo.")

    def agregar_dinero(self, cantidad):#añadir saldo a la cuenta
        if cantidad <= 0:
            raise ValueError("La cantidad que vayas a ingresar tiene que ser positiva.")
        self.saldo += cantidad
        print(f"{self.nombre} has ingresado {cantidad}€. Tu saldo actual es: {self.saldo}€")

    def transferir_dinero(self, destino, cantidad):#hacer transferencia
        if not self.tiene_cuenta_corriente or not destino.tiene_cuenta_corriente:
            raise ValueError("Si tú o el otro usuario no tenéis cuenta no se puede hacer transferencia.")
        if cantidad <= 0:
            raise ValueError("La cantidad a transferir debe ser positiva.")
        if cantidad > self.saldo:
            raise ValueError(f"{self.nombre} no tiene saldo suficiente para realizar la transferencia {cantidad}€.")

        self.saldo -= cantidad
        destino.saldo += cantidad
        print(f"{self.nombre} ha transferido {cantidad}€ a {destino.nombre}.")
        print(f"Nuevo saldo de {self.nombre}: {self.saldo}€")
        print(f"Nuevo saldo de {destino.nombre}: {destino.saldo}€")


alicia = UsuarioBanco("Eva", 1000, True)
bob = UsuarioBanco("Noelia", 500, True)

bob.agregar_dinero(120)

bob.transferir_dinero(alicia, 60)


alicia.retirar_dinero(50)
```

Kata 36
Crea una función llamada procesar_texto
Procesa un texto según la opción especificada: contar_palabras, reemplazar_palabras o eliminar_palabra.
Código a seguir:
Crear una función contar_palabras que cuente el número de veces que aparece cada palabra en el texto y devuelva un diccionario.
Crear una función reemplazar_palabras para sustituir una palabra_original por una palabra_nueva en el texto y devolver el texto modificado.
Crear una función eliminar_palabra que elimine una palabra del texto y devuelva el texto sin ella.
Crear la función procesar_texto que reciba un texto, una opción ("contar", "reemplazar", "eliminar") y un número variable de argumentos según la opción elegida.
Caso de uso:
Verificar el funcionamiento completo de procesar_texto.

```python
def contar_palabras(texto):
    palabras = texto.split()  # Dividimos el texto en palabras
    contador = {}  # Diccionario vacío para almacenar conteos

    for palabra in palabras:
        palabra = palabra.lower().strip(".,;:!?")  # Limpiamos puntuación y pasamos a minúsculas
        contador[palabra] = contador.get(palabra, 0) + 1  # Sumamos una ocurrencia

    return contador


def reemplazar_palabras(texto, palabra_original, palabra_nueva):#se sustituye la palabra por otra dentro del texto
  
    return texto.replace(palabra_original, palabra_nueva)


def eliminar_palabra(texto, palabra):#eliminamos la palabra
    palabras = texto.split()
    nuevas_palabras = [p for p in palabras if p.lower() != palabra.lower()]
    return " ".join(nuevas_palabras)


def procesar_texto(texto, opcion, *args):
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        if len(args) != 2:
            raise ValueError("Debes proporcionar palabra_original y palabra_nueva.")
        return reemplazar_palabras(texto, args[0], args[1])
    elif opcion == "eliminar":
        if len(args) != 1:
            raise ValueError("Debes indicar la palabra a eliminar.")
        return eliminar_palabra(texto, args[0])
    else:
        raise ValueError("Opción no válida. Usa 'contar', 'reemplazar' o 'eliminar'.")


texto_prueba = "El sol brilla sobre el mar y el mar refleja el sol"

print("== TEXTO ORIGINAL ==")
print(texto_prueba)
print("\n")

print("Resultado del conteo de palabras:")
print(procesar_texto(texto_prueba, "contar"))

print("\nReemplazando 'sol' por 'luna':")
print(procesar_texto(texto_prueba, "reemplazar", "sol", "luna"))


print("\nEliminando 'mar':")
print(procesar_texto(texto_prueba, "eliminar", "mar"))
```

## Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo", "circulo" o "triangulo") y datos (una tupla con los datos necesarios para calcular el área de la figura).

```python


import math  # Importamos math para poder usar pi en el área del círculo

def calcular_area(figura, datos):
    figura = figura.lower()  # Pasamos a minúsculas para evitar errores por mayúsculas, condicionales segun figura geométrica

    
    if figura == "rectangulo":
        base, altura = datos  # sacamos datos de la tupla
        area = base * altura
        return f"El área del rectángulo es: {area} unidades cuadradas"

    elif figura == "circulo":
        (radio,) = datos  # tupla con un solo valor
        area = math.pi * radio**2
        return f"El área del círculo es: {area:.2f} unidades cuadradas"

    elif figura == "triangulo":
        base, altura = datos
        area = (base * altura) / 2
        return f"El área del triángulo es: {area} centímetros cuadrados"

    else:
        raise ValueError("Figura no reconocida. Usa 'rectangulo', 'circulo' o 'triangulo'.")

try:
    print(calcular_area("rectangulo", (8, 5)))   
    print(calcular_area("circulo", (6,)))        
    print(calcular_area("triangulo", (10, 7)))   

except ValueError as e:
    print("ERROR:", e)

finally:
    print("=== Programa finalizado ===")
```
